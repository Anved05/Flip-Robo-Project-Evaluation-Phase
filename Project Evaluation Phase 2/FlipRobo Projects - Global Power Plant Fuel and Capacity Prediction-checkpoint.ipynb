# Preprocessing

```python
# Dropping the data according to point no. 1
df=df.drop(['country','country_long','name','wepp_id','generation_gwh_2013','generation_gwh_2019','estimated_generation_gwh','gppd_idnr','other_fuel1','other_fuel2','other_fuel3','url','generation_data_source','owner'],axis=1)
df
```

```python
# Imputing the data in the columns
# Imputing latitute by mode as that cannot be imputed through mean
df['latitude']=df['latitude'].fillna(df['latitude'].mode()[0])

# Imputing longitude by mode as that cannot be imputed through mean
df['longitude']=df['longitude'].fillna(df['longitude'].mode()[0])

# Imputing Capacity in megawatt by mean
df['capacity_mw']=df['capacity_mw'].fillna(df['capacity_mw'].mean())

# Filling the commissioning year with mode
df['commissioning_year']=df['commissioning_year'].fillna(df['commissioning_year'].mode()[0])

# Filling the year of capacity with mode
df['year_of_capacity_data']=df['year_of_capacity_data'].fillna(df['year_of_capacity_data'].mode()[0])

# Imputing generation_gwh_2014 by mean
df['generation_gwh_2014']=df['generation_gwh_2014'].fillna(df['generation_gwh_2014'].mean())

# Imputing generation_gwh_2015 by mean
df['generation_gwh_2015']=df['generation_gwh_2015'].fillna(df['generation_gwh_2015'].mean())

# Imputing generation_gwh_2016 by mean
df['generation_gwh_2016']=df['generation_gwh_2016'].fillna(df['generation_gwh_2016'].mean())

# Imputing generation_gwh_2017 by mean
df['generation_gwh_2017']=df['generation_gwh_2017'].fillna(df['generation_gwh_2017'].mean())

# Imputing generation_gwh_2018 by mean
df['generation_gwh_2018']=df['generation_gwh_2018'].fillna(df['generation_gwh_2018'].mean())

# Filling geolocation_source with mode
df['geolocation_source']=df['geolocation_source'].fillna(df['geolocation_source'].mode()[0])
```

# Regression model for the Prediction of Megawat power

```python
#Splitting the datbase into inputs and output
x=df.drop(['capacity_mw'],axis=1)
y=df['capacity_mw']
print(x.shape)
print(y.shape)
```

```python
# Removing Skewness Using Power Transform
from sklearn.preprocessing import power_transform
x=power_transform(x, method='yeo-johnson')
x
```

```python
#checking for skewness is removed or not
skew_df=pd.DataFrame(x)
skew_df.skew()
```

We cannot further remove skewness hence proceeding with this only

```python
#importing the StandardScaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
x
```

```python
# importing Linear Regression and metrics for it
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# Finding the best Random State
maxr2=0
maxrs=0
lr=LinearRegression()
for i in range(0,200):
    print("Random State is ", i)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(xtrain,ytrain)
    predtrain=lr.predict(xtrain)
    predtest=lr.predict(xtest)
    print('training accuracy is ', r2_score(ytrain,predtrain))
    score=r2_score(ytest,predtest)
    print('testing accuracy is ', score)
    print("\n")
    
    if score>maxr2:
        maxr2=score
        maxrs=i
        
        
print("maximum r2 score is ", maxr2 , "at random state ", maxrs)
```

We have got the best random state as 61 where the testing accuracy is 0.59 and training is 0.42

```python
# Spliting data on the best Random state we have got
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=61)
```

```python
#running that same model again for other metrics too

lr.fit(x_train,y_train)
predtrainlr=lr.predict(x_train)
predtestlr=lr.predict(x_test)
print('training accuracy is ', r2_score(y_train,predtrainlr))
print('testing accuracy is ', r2_score(y_test,predtestlr))
print("Mean Squared Error is ",mean_squared_error(y_test,predtestlr))
print("Mean Absolute Error is ",mean_absolute_error(y_test,predtestlr))
print("Root Mean Squared Error is ",np.sqrt(mean_squared_error(y_test,predtestlr)))
```

```python
from sklearn.model_selection import cross_val_score
#Cross validation of the lr model
for j in range(2,20):
    cv_score=cross_val_score(lr, x, y, cv=j)
    print('At fold ', j , "Mean CV score is ", cv_score.mean())
    print('training accuracy is ', r2_score(y_train,predtrainlr))
    print('testing accuracy is ', r2_score(y_test,predtestlr))
```

```python
# modelling using the Decision tree regressor
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
predtraindtr=dtr.predict(x_train)
predtestdtr=dtr.predict(x_test)
print('training accuracy is ', r2_score(y_train,predtraindtr))
print('testing accuracy is ', r2_score(y_test,predtestdtr))
print("Mean Squared Error is ",mean_squared_error(y_test,predtestdtr))
print("Mean Absolute Error is ",mean_absolute_error(y_test,predtestdtr))
print("Root Mean Squared Error is ",np.sqrt(mean_squared_error(y_test,predtestdtr)))
```

```python
import yellowbrick
from  yellowbrick.classifier import ROCAUC

def plot_ROC_curve(model, X_train, Y_train, X_test, Y_test):

    # Creating visualization with the readable labels
    visualizer = ROCAUC(model, encoder={0: 'Biomass', 1: 'Coal', 2: 'Gas',3:'Hydro',4:'Nuclear',5:'Oil',6:'Solar',7:'Wind'})
                                        
    # Fitting to the training data first then scoring with the test data                                    
    visualizer.fit(X_train, Y_train)
    visualizer.score(X_test, Y_test)
    visualizer.show()
    
    return visualizer
```
